```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
data = load_digits()
dir(data)
```

```python
data.images[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.matshow(data.images[0])
```

```python
data
```

```python
df = pd.DataFrame(data.data)
df.head()
```

```python
data.target
```

```python
df['target'] = data.target
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis = 'columns')
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.20)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# model = RandomForestClassifier() you can tune your model to get better score
model = RandomForestClassifier(n_estimators= 55)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

## Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## Exercise

#### Use famous iris flower dataset from sklearn.datasets to predict flower species using random forest classifier.
  1. Measure prediction score using default n_estimators (10)
  2. Now fine tune your model by changing number of trees in your classifer and tell me what best score you can get using how many trees

```python
from sklearn.datasets import load_iris
```

```python
iris  = load_iris()
iris
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
# feature names are basically the columns name
# target names are basically the target columns name 
df.head()
```

```python
df['target'] = iris.target
df
```

```python
X = df.drop('target', axis = 'columns') 
y = df.target

# X = df.iloc[:,0:4]
# y = df.iloc[:,4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators= 10)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# extra added by me
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm, annot= True)
plt.ylabel('Truth')
plt.xlabel('Predicted')
```
